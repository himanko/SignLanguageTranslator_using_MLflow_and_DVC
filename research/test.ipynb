```python
import os
import shutil
import random

# Define the path to the root folder
root_folder = r"C:\\Users\\himan\\CollegeProject\\datasets\WLASL\\landmarks_files_(alphabatically)\\WLASL_300 - Copy"
moved_dest_root = r"C:\\Users\\himan\\CollegeProject\\datasets\WLASL\\landmarks_files_(alphabatically)\\WLASL-300-val"

# Initialize a dictionary to store the counts for each class_name
video_counts = {}

# Function to count videos in a directory
def count_videos(directory):
    video_count = 0
    for file in os.listdir(directory):
        if file.endswith(".mp4") or file.endswith(".avi") or file.endswith(".parquet"):
            video_count += 1
    return video_count

# Function to move 20% of videos from a folder to another folder
def move_videos(class_path, dest_folder):
    files_to_move = os.listdir(class_path)
    num_files_to_move = int(len(files_to_move) * 0.2)  # 20% of files
    files_to_move = random.sample(files_to_move, num_files_to_move)
    for file_name in files_to_move:
        source = os.path.join(class_path, file_name)
        destination = os.path.join(dest_folder, file_name)
        shutil.move(source, destination)

# Traverse through each class_name folder and count videos
for class_name in os.listdir(root_folder):
    class_path = os.path.join(root_folder, class_name)
    if os.path.isdir(class_path):
        video_counts[class_name] = count_videos(class_path)
        # Create a destination folder for moved videos
        moved_dest_folder = os.path.join(moved_dest_root, class_name)
        os.makedirs(moved_dest_folder, exist_ok=True)
        # Move 20% of videos to the destination folder
        move_videos(class_path, moved_dest_folder)

# Print the results
for class_name, count in video_counts.items():
    print(f"Class {class_name}: {count} videos")

# # Traverse through each class_name folder and count videos
# for class_name in os.listdir(moved_dest_root):
#     class_path = os.path.join(moved_dest_root, class_name)
#     if os.path.isdir(class_path):
#         video_counts[class_name] = count_videos(class_path)

# # Print the results
# for class_name, count in video_counts.items():
#     print(f"Class {class_name}: {count} videos")
```
